# Dealing with html tags

```python
import re 
def remove_html_tags(text):
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',text)
```

```python
text='<p>Hi hello</p> <h1>single seena </h1>'
```

```python
remove_html_tags(text)
```

# Removing url in the text

```python
def remove_url(text):
    pattern=re.compile('https?://\S+|WWW\.\S+')
    return pattern.sub(r'',text)
```

```python
text='https://www.hackerrank.com/profile/srinivasdr25 hi hello single seena'
```

```python
remove_url(text)
```

```python
nlp.head()
```

```python
nlp['Message_body'] = nlp['Message_body'].apply(remove_url)
```
